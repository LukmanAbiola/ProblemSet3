```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
roman_emp_data = pd.read_csv("roman-emperor-reigns.csv")
print(roman_emp_data.head())
```

```python
assassinated_emp_count = roman_emp_data[roman_emp_data['Cause_of_Death'].isin(["Assassinated","Possibly assassinated"])].shape[0]

non_assassinated_emp_count = roman_emp_data[(~roman_emp_data['Cause_of_Death'].isin(["Assassinated","Possibly assassinated"]))].shape[0]


pie_labels = ['Assassinated/Possibly Assassinated', 'Others']
count_list = [assassinated_emp_count, non_assassinated_emp_count]

# plotting
plt.pie(count_list , labels = pie_labels, autopct = '%1.2f%%', startangle = 80)
plt.axis('equal')
plt.title("fraction of all Roman Emperors that were assassinated/possibly assassinated")
plt.show()
```
