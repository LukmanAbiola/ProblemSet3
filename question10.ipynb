```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python

revenue_phd_df = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")

colors_list = ["red", "green", "blue", "orange", "purple", "cyan", "magenta", "yellow", "brown", "pink"]
color_map = ListedColormap(colors_list)

plt.scatter(revenue_phd_df["Total Arcade Revenue (billions)"], revenue_phd_df["Computer Science Doctorates Awarded (US)"], c = revenue_phd_df["Year"], cmap=color_map)

color_bar = plt.colorbar()
color_bar.set_label("Year")

plt.xlabel("Total Arcade Revenue (in Billions)")
plt.ylabel("Computer Science PhDs awarded in the U.S")
plt.title("Relationship between revenue from Arcades and the number of Computer Science PhDs between 2000-2009")

plt.show()
```
