```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
marriage_divorce_data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
print(marriage_divorce_data.head())
```

```python
years = [1900, 1950, 2000]
selected_data = marriage_divorce_data[marriage_divorce_data["Year"].apply(lambda x: x in years)]
plt.figure(figsize=(8, 6))

plt.bar(selected_data["Year"] - 4, selected_data["Marriages_per_1000"], width=8, label="Marriages", color = "lightgreen")
plt.bar(selected_data["Year"] + 4, selected_data["Divorces_per_1000"], width=8, label="Divorces" , color = "red")

plt.xlabel("Year")
plt.ylabel("Per Capita")
plt.title("Number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.")
plt.xticks(years)
plt.legend()
plt.show()
```
